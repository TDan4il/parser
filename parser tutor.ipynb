```python
# Указываем откуда парсим
url = "https://tutortop.ru/courses_category/programmirovanie/"

headers = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36 OPR/86.0.4363.59"
}

# Get запрос, получаем текст страницы
req = requests.get(url, headers=headers)
src = req.text
```

Сохраняем для последующей работы + чтобы не беспокоить сайт постоянными запросами

```python
with open("index2.html", "w") as file:
    file.write(src)

with open("index2.html") as file:
    src = file.read()
```

Ищем обучающие программы

```python
soup = BeautifulSoup(src, "lxml")
course = soup.find_all("span", class_="scw-name")
data_course = []

for i in course:
    cs = i.text
    print(cs)
    data_course.append(cs)

print(len(data_course))
#print(len(course2))
```

```python
with open("/content/10 лучших онлайн-курсов программирования с нуля 2022.html") as file:
    src = file.read()
```

```python
soup = BeautifulSoup(src, "lxml")
#course = soup.find_all("span", class_="scw-name")
course = soup.find_all("div", class_="m-course-title")
data_course = []

for i in course:
    cs = i.text
    print(cs)
    data_course.append(cs)

print(len(data_course))
#print(len(course2))
```

```python
i = 0
while i < len(data_course):
  #print(i)
  if data_course[i][0] == " ":
    #print(data_dlitelnost[i])
    del data_course[i]
  else: i += 1

print(data_course)
len(data_course)
```

# Парсинг

Ищем школы

```python
school = soup.find_all(class_="course__wrap__box__top_school_name")
data_school = []

for i in school:
    sc = i.text
    print(sc)
    data_school.append(sc)

len(data_school)
```

Длительность обучения

```python
dlitelnost = soup.find_all("div", class_="tab-course-col tab-course-col-flex tab-course-col-dlitelnost initial-version")
data_dlitelnost = []

for i in dlitelnost:
    dl = i.text
    print(dl)
    data_dlitelnost.append(dl)

len(data_dlitelnost)
```

Поскольку в data_dlitelnost указаны и бесплатные курсы, убираем их (можно заметить что бесплатные курсы начинаются с отступа). 

```python
#for i in range(0, len(data_dlitelnost)):
i = 0
while i < len(data_dlitelnost):
  #print(i)
  if data_dlitelnost[i][0] == " ":
    #print(data_dlitelnost[i])
    del data_dlitelnost[i]
  else: i += 1

print(data_dlitelnost)
len(data_dlitelnost)
```

Ищем цены

```python
price = soup.find_all("div", class_="tab-course-col tab-course-col-flex tab-course-col-price initial-version")
data_price = []

for i in price:
    pr = i.text
    print(pr)
    data_price.append(pr)

len(data_price)
```

Собираем описания

```python
features = soup.find_all("div", class_="course__wrap__box__content")
data_features = []

for i in features:
    ft = i.text
    print(ft)
    data_features.append(ft)

len(data_features)
```

Создаем csv файл

```python
d = {
    "Курс": pd.Series(data_course),
    "Школа": pd.Series(data_school),
    "Срок обучения": pd.Series(data_dlitelnost),
    "Цена": pd.Series(data_price),
    "Содержание курса": pd.Series(data_features),
}

df = pd.DataFrame(d)
df
```

```python
df.to_csv("Tutor.csv")
```
